```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
import math
import cmath
import pandas as pd
#from scipy.optimize import fsolve
#from scipy.optimize import curve_fit
#from matplotlib.widgets import Slider, Button
#from lmfit import minimize, Parameters, fit_report
```

```python
L = [1]
L.pop()

for i in range(5):
    file="F1P" + str(5*(i+1)) + ".csv"

    names=['Freq(GHz)', 'Amp(dB)']
    df=pd.read_csv(file, names=names, skiprows=18, nrows=100)
    names2=['Freq(GHz)', 'Phase(rad)']
    df2=pd.read_csv(file, names=names2, skiprows=125, nrows=100)

    Freq=df['Freq(GHz)']/1e9
    Amp=df['Amp(dB)']+65

    Freq2=df2['Freq(GHz)']/1e9
    Pha=df2['Phase(rad)']
    df2
    
    L.append(Freq)
    L.append(Amp)
    L.append(Pha)
    
    file="F1P" + str(50*i+75) + ".csv"

    names=['Freq(GHz)', 'Amp(dB)']
    df=pd.read_csv(file, names=names, skiprows=18, nrows=100)
    names2=['Freq(GHz)', 'Phase(rad)']
    df2=pd.read_csv(file, names=names2, skiprows=125, nrows=100)

    Freq=df['Freq(GHz)']/1e9
    Amp=df['Amp(dB)']+65

    Freq2=df2['Freq(GHz)']/1e9
    Pha=df2['Phase(rad)']
    df2
    
    L.append(Freq)
    L.append(Amp)
    L.append(Pha)
    
file="F1P30.csv"

names=['Freq(GHz)', 'Amp(dB)']
df=pd.read_csv(file, names=names, skiprows=18, nrows=100)
names2=['Freq(GHz)', 'Phase(rad)']
df2=pd.read_csv(file, names=names2, skiprows=125, nrows=100)

Freq=df['Freq(GHz)']/1e9
Amp=df['Amp(dB)']+65
Freq2=df2['Freq(GHz)']/1e9
Pha=df2['Phase(rad)']

L.append(Freq)
L.append(Amp)
L.append(Pha)
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,4))
for i in range(11):
    ax[0].plot(L[i*3],L[i*3+2])
    ax[1].plot(L[i*3],L[i*3+1])
    
titl = ["-5 dBm"]
for i in range(5):
    titl.append("-" + str(7.5+5*i) + " dBm")
    titl.append("-" + str(10+5*i) + " dBm")

ax[0].legend(titl, loc="lower right", fontsize = "x-small")
ax[1].legend(titl, fontsize="x-small")

ax[0].set_xlabel('Frequency')
ax[1].set_xlabel('Frequency')
ax[0].set_ylabel('Phase (rad)')
ax[1].set_ylabel(r'$S_{21}(db)$')

plt.show()


```
