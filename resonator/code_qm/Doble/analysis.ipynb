```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib tk
```

```python
### f is float in GHz
from scipy.constants import c
def leng(f):
    f2=f*1e9 ## to GHz
    eff=(11.9+1)/2
    le=c/(4*np.sqrt(eff)*f2)
    le=le*1e3 # to milimeters
    return le
leng(5.0004)
```

```python
df=pd.read_csv('data_analysis.txt', skiprows=15, sep='\s+')
fig,ax=plt.subplots()
ax.plot(1/df['LongB(mm)'],df['FreqB(GHz)'],'-o', label='FreqB')
ax.plot(1/df['LongB(mm)'], df['FreqA(GHz)'],'-o', label='FreqA')
ax.plot(1/leng(df['Sim']),df['Sim'], '--', label='f_B')
ax.plot(1/leng(df['Sim']),df['FreA'], '--', label='f_A')
ax.set_xlabel('$1/L_B$')
ax.set_ylabel('$f_0$')
plt.legend()
```
